```python
import os
data_dir = './garbage_data/'
# Use !ls {data_dir} to check the structure if needed
```

```python
IMG_SIZE = (160, 160) # Or another size suitable for your chosen model
BATCH_SIZE = 32
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Rescale all pixel values to be between 0 and 1
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Create the training dataset generator
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical', # Use 'categorical' for multi-class classification
    subset='training'
)

# Create the validation dataset generator
validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# Get the number of classes from the generator
num_classes = train_generator.num_classes

# 1. Load the base model (pre-trained on ImageNet)
base_model = MobileNetV2(
    input_shape=IMG_SIZE + (3,), # Add 3 for RGB channels
    include_top=False,           # Exclude the classification layer of the original model
    weights='imagenet'
)

# 2. Freeze the base model layers
base_model.trainable = False

# 3. Add new classification layers (the 'head')
x = base_model.output
x = GlobalAveragePooling2D()(x) # Collapse the 3D feature map to a 1D vector
x = Dense(128, activation='relu')(x)
predictions = Dense(num_classes, activation='softmax')(x) # Final output layer

# 4. Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# 5. Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
EPOCHS = 10 # Start with a small number of epochs

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    verbose=1
)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define Model Parameters (as before)
IMG_SIZE = (160, 160)
BATCH_SIZE = 32

# CORRECTED Data Directory
data_dir = './garbage_data/garbage-dataset/'

# Rescale all pixel values to be between 0 and 1
# We will use the validation_split to create train/validation sets from the single folder
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Create the training dataset generator
print("Loading Training Data...")
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

# Create the validation dataset generator
print("Loading Validation Data...")
validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# Get the correct number of classes from the generator
num_classes = train_generator.num_classes

# 1. Load the base model (MobileNetV2 is excellent for Edge AI)
base_model = MobileNetV2(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights='imagenet'
)

# 2. Freeze the base layers
base_model.trainable = False

# 3. Add your classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
predictions = Dense(num_classes, activation='softmax')(x)

# 4. Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# 5. Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
EPOCHS = 10 # You can adjust this number
print("Starting Model Training...")

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    verbose=1
)

# Save the model once training is complete
model.save("garbage_classifier.h5")
print("Model trained and saved as garbage_classifier.h5")
```

```python
import numpy as np

# Load the trained Keras model
model = tf.keras.models.load_model("garbage_classifier.h5")

# 1. Corrected Representative Dataset Generator for Quantization Calibration
def representative_data_gen():
  # We use the ITERATOR from the generator
  data_iterator = iter(train_generator)

  # Collect 10 batches of data (adjust this number if needed)
  for i in range(10):
    # Get the next batch: (images, labels)
    try:
      input_value = next(data_iterator)
      # We only need the images (index 0) and convert them to float32
      # Note: TFLite inference expects float32 OR int8 depending on conversion settings
      yield [input_value[0].astype(np.float32)]
    except StopIteration:
      # Stop if the iterator runs out of data before 10 batches
      print(f"Reached end of generator after {i} batches.")
      break

# 2. Set up the TFLite Converter for Full Integer Quantization
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = representative_data_gen

# Crucial for Edge AI: Force Integer Input/Output types
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8

# 3. Convert and Save the TFLite Model
print("Starting TFLite conversion and quantization...")
tflite_model = converter.convert()

with open('garbage_classifier_quantized.tflite', 'wb') as f:
  f.write(tflite_model)

print("\n✅ TFLite model successfully saved: garbage_classifier_quantized.tflite")
```

```python
import numpy as np
import tensorflow as tf
from tqdm import tqdm
import math

# --- Setup ---
# 1. Load the TFLite model and allocate tensors.
interpreter = tf.lite.Interpreter(model_path='garbage_classifier_quantized.tflite')
interpreter.allocate_tensors()

# 2. Get input and output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
input_index = input_details[0]['index']
output_index = output_details[0]['index']

# TFLite Model expects shape: (1, 160, 160, 3)
IMG_SIZE = tuple(input_details[0]['shape'][1:3])
input_dtype = input_details[0]['dtype']
input_scale, input_zero_point = input_details[0]['quantization']

correct_predictions = 0
total_samples = 0

print("Starting TFLite model evaluation (processing image by image)...")

# --- Evaluation Loop ---
# We will iterate through ALL samples, one at a time.
num_samples = validation_generator.samples
num_batches = math.ceil(num_samples / validation_generator.batch_size) # Use ceil for final incomplete batch

# Reset the generator to ensure we start from the beginning of the validation set
validation_generator.reset()

for i in tqdm(range(num_batches)):
    # Get a batch of data (images, labels)
    val_images_batch, val_labels_batch = next(validation_generator)

    # Iterate through each image in the current batch
    for j in range(val_images_batch.shape[0]):
        # Get a single image and expand its dimensions to (1, H, W, C)
        single_image = val_images_batch[j:j+1] # Slice [j:j+1] keeps the batch dimension of 1
        true_class = np.argmax(val_labels_batch[j]) # True class index

        # 3. Preprocess the image for the TFLite model (Quantization specific)

        # Scale float data (0-1) back to 0-255
        scaled_image = single_image * 255.0

        # Quantize the float data to int8
        quantized_image = (scaled_image / input_scale) + input_zero_point
        quantized_image = quantized_image.astype(input_dtype)

        # 4. Run Inference

        # NOTE: We DO NOT need to resize_tensor_input here, as the shape (1, 160, 160, 3)
        # is constant and expected by the interpreter.

        interpreter.set_tensor(input_index, quantized_image)
        interpreter.invoke()

        # 5. Post-process and Check Prediction
        tflite_output = interpreter.get_tensor(output_index)

        # Dequantize output if necessary
        if output_details[0]['dtype'] == np.int8:
            output_scale, output_zero_point = output_details[0]['quantization']
            tflite_output = (tflite_output.astype(np.float32) - output_zero_point) * output_scale

        predicted_class = np.argmax(tflite_output[0]) # Prediction is at index 0 because batch size is 1

        # Count correct predictions
        if predicted_class == true_class:
            correct_predictions += 1
        total_samples += 1

# Final Accuracy Calculation
tflite_accuracy = correct_predictions / total_samples
print(f"\nOriginal Keras Model Total Validation Samples: {num_samples}")
print(f"TFLite Model Total Correct Predictions: {correct_predictions}")
print(f"**Final TFLite Validation Accuracy: {tflite_accuracy:.4f}**")
```
